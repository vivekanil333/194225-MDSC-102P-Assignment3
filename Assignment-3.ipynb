$Binomial  distribution$ 

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
def B(n,p):
    f=lambda r,n,p: np.math.comb(n,r) *((p)**r)*((1-p)**(n-r))
    r=np.arange(0,n,dtype=int)
    y=[f(i,n,p) for i in r]
    plt.title("Binomial Distribution")
    plt.xlabel("x-values")
    plt.ylabel("f(x)")
    plt.plot(r, y)
    plt.show()
    
B(40,0.4)



    
    




    


    
    
```

$poisson distribution$

```python
def Poi(a):
    f = lambda x,a: np.exp(-a) *((a**x) / (np.math.factorial(x)))
    x = np.arange(0, 30, dtype=int)
    y = [f(i,a) for i in x]
    plt.title("Poisson Distribution")
    plt.xlabel("x-values")
    plt.ylabel("f(x)")
    plt.plot(x, y)
    plt.show()
    
Poi(0.03)
```

$exponential distribution$

```python
def Exp(a):
    f = lambda x,a: a * np.exp((-a) * x)
    x = np.arange(0, 30, dtype=int)
    y = [f(i,a) for i in x]
    plt.title("Exponential Distribution")
    plt.xlabel("x-values")
    plt.ylabel("f(x)")
    plt.plot(x, y)
    plt.show()
Exp(2)
```

$normal distribution$

```python
def N(μ, var):
    f = lambda x,μ,var: (1/(np.sqrt(2*np.pi*var))) * np.exp(-0.5*(((x-μ)**2)/var))
    x = np.linspace(-20, 10, 100)
    y = [f(i,μ,var) for i in x]
    plt.title("Normal Distribution")
    plt.xlabel("x-values")
    plt.ylabel("f(x)")
    plt.plot(x, y)

    μ = 2 
    var = 10
    y = [f(i,μ,var) for i in x]
    plt.plot(x, y)

    μ = 7
    var = 7
    y = [f(i,μ,var) for i in x]
    plt.plot(x, y)

    plt.legend([" μ = 0, var = 1", "μ = 2, var = 10", "μ = 7, var = 71"])
    plt.show()

N(0,2)
```

$Gamma Distribution$

```python
def Gamma(alpha, lam):
    f = lambda x,alpha,lam: ((lam**alpha)/np.math.factorial(alpha)) * x**(lam - 1) * np.exp((-lam) * x)
    x = np.linspace(0, 10, 100)
    y = [f(i, alpha, lam) for i in x]
    plt.title("Gamma Distribution")
    plt.xlabel("x-values")
    plt.ylabel("f(x)")
    plt.plot(x, y)

    alpha = 4
    lam = 2
    y = [f(i, alpha, lam) for i in x]
    plt.plot(x, y)

    alpha = 3
    lam = 1
    y = [f(i, alpha, lam) for i in x]
    plt.plot(x, y)

    plt.legend(["alpha = 5, lam = 3", "alpha = 4, lam = 2", "alpha = 3, lam = 1"])
    plt.show()

Gamma(5,3)
```
